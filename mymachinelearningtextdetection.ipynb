```
def draw_boxes(image,r,color='red',width=0):
    draw = ImageDraw.Draw(image)
    for re in r:
        p0,p1,p2,p3 = re[0]
        draw.line([*p0,*p1,*p2,*p3],fill=color,width=width)
    return image
draw_boxes(ime,r)        
```
