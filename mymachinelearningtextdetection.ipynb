All the package to implenment textdetectoin

```
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
import PIL
from PIL import ImageDraw
```

CREATING READER TO DETECTION OF ENGLISH AND HINDI LANGUAGE

```
reader = easyocr.Reader(['en','hi'])
```

PROCESS THE IMAGE WHICH HAS ENGLISH WORD'S

```
ime = PIL.Image.open("eimage.jpg")
ime
```

PROCESS THE IMAGE WHICH HAS HINDI WORD'S

```
imh = PIL.Image.open("himage.jpg")
imh
```

DETECTION OF ALL TEXT IN ENGLISH IMAGE DISPLAY 

```
ren = reader.readtext("eimage.jpg")
ren
```

DETECTION OF ALL TEXT IN HINDI IMAGE DISPLAY 

```
rhi = reader.readtext("himage.jpg")
rhi
```

DETECTION OF TEXT IN ENGLISH IMAGE AND GIVE BORDER WITH RED COLOR

```
def draw_boxes(image,r,color='red',width=0):
    draw = ImageDraw.Draw(image)
    for re in r:
        p0,p1,p2,p3 = re[0]
        draw.line([*p0,*p1,*p2,*p3],fill=color,width=width)
    return image
draw_boxes(ime,ren)        
```

DETECTION OF TEXT IN HINDI IMAGE AND GIVE BORDER WITH RED COLOR

```
draw_boxes(imh,rhi)
```

DISPLAY ALL THE TEXT DETECTED IN ENGLISH IMAGE

```
for i in ren:
    print(i[1])
```

DISPLAY ALL THE TEXT DETECTED IN HINDI IMAGE

```
for x in rhi:
    print(x[1])
```
